Write a program to plot Sigmoid activation functions that are being used in neural
networks.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x=np.linspace(-5,5,100)
```

```python
plt.plot(x,1/(1+np.exp(-x)),label='Sigmoid')
plt.xlabel('Input')
plt.ylabel('Activation')
plt.title('Activation Function')
plt.legend()
plt.grid(True)
```
