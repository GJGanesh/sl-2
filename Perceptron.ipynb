With a suitable example demonstrate the perceptron learning law with its decision
regions using python. Give the output in graphical form.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.array([[0, 0], [1, 0], [0, 1], [1, 1]])
Y = np.array([-1, -1, -1, 1])
```

```python
w = np.zeros(X.shape[1])
b = 0
```

```python
for _ in range(6):
    for i in range(X.shape[0]):
        y_pred = np.sign(np.dot(X[i], w) + b)
        
        if y_pred != Y[i]:
            w += 0.3 * Y[i] * X[i]
            b += 0.3 * Y[i]
```

```python
x_min, x_max = X[:, 0].min() - 1,  X[:, 0].max() + 1
y_min, y_max = X[:, 1].min() - 1,  X[:, 1].max() + 1
xx, yy = np.meshgrid(np.arange(x_min, x_max, 0.01),np.arange(y_min, y_max, 0.01))
```

```python
Z = np.sign(np.dot(np.c_[xx.ravel(), yy.ravel()], w) + b)
Z = Z.reshape(xx.shape)
```

```python
plt.contourf(xx, yy, Z, alpha=0.5)
plt.scatter(X[:, 0], X[:, 1], c=Y)
plt.xlabel('X1')
plt.ylabel('X2')
plt.title('Perceptron Decision Regions')
plt.show()
```
